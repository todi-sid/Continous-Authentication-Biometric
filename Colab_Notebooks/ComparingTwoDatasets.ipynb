```python
dataset_1=pd.read_csv(path+"file_1.csv")
#parse strings to datetime type
dataset_1['time']=pd.to_datetime(dataset_1['time'],infer_datetime_format=True)
indexedDataset_1=dataset_1.set_index(['time'])
```

```python
dataset_2=pd.read_csv(path+"file_2.csv")
#parse strings to datetime type
dataset_2['time']=pd.to_datetime(dataset_2['time'],infer_datetime_format=True)
indexedDataset_2=dataset_2.set_index(['time'])
```

```python
##plot graph 1
plt.xlabel("Time")
plt.ylabel("HeartRate")
plt.plot(indexedDataset_1)
```

```python
##plot graph 2
plt.xlabel("Time")
plt.ylabel("HeartRate")
plt.plot(indexedDataset_2)
```

```python
##plot graph
plt.xlabel("Time")
plt.ylabel("HeartRate")
plt.plot(indexedDataset_1,color = 'red',label='Data 1')
plt.plot(indexedDataset_2, color = 'blue',label='Data 2')
plt.legend()
```

```python
#Determining rolling statistics 1
rolmean_1=indexedDataset_1.rolling(window=10).mean()
rolstd_1=indexedDataset_1.rolling(window=10).std()
```

```python
#Determining rolling statistics
rolmean_2=indexedDataset_2.rolling(window=10).mean()
rolstd_2=indexedDataset_2.rolling(window=10).std()
```

```python
print(rolmean_1,rolmean_2)
```

```python
print(rolstd_1,rolstd_2)
```

```python
#Plot rolling statistics
plt.plot(rolmean_1,color='red',label='Rolling mean_1')
plt.plot(rolmean_2,color='blue',label='Rolling mean_2')
plt.legend(loc='best')
plt.title('Rolling mean')
plt.show(block=False)
```

```python
#Plot rolling statistics
plt.plot(rolstd_1,color='red',label='Rolling std_1')
plt.plot(rolstd_2,color='blue',label='Rolling std_2')
plt.legend(loc='best')
plt.title('Rolling Standard Deviation')
plt.show(block=False)
```
